# Trajectory of a Ball on Earth (Gravity)

### The equations of motion for the trajectory of a ball launched under gravity are given by :

##### Horizontal motion:
##### $x = v_{0}x \times t$

##### Vertical motion:
##### $y = v_{0}y \times t - \frac{1}{2} \times g \times t^2$

##### Where:
1. $x$ is the horizontal distance traveled by the ball,
2. $y$ is the vertical height above the launch point,
3. $t$ is the time of flight,
4. $v_{0}x$ = $v_{0} \times cos(angle)$ is the initial horizontal velocity,
5. $v_{0}y$ = $v_{0} \times sin(angle)$ is the initial vertical velocity,
6. $g$ is the acceleration due to gravity.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def projectile_motion(v0, angle_degrees, g=9.8106, num_steps=100):
    # Converting angle from degrees to radians
    angle_radians = np.radians(angle_degrees)
    # Calculating components of initial velocity
    v0x = v0 * np.cos(angle_radians)
    v0y = v0 * np.sin(angle_radians)
    
    # Calculating time steps and time interval
    total_time = (2 * v0y) / g
    time_interval = total_time / num_steps

    # Initializing lists to store position data
    x_list, y_list = [], []

    # For each time sample we calculate the x and y coordinate of trajectory
    for i in range(num_steps + 1):
        t = i * time_interval
        x = v0x * t
        y = v0y * t - 0.5 * g * t**2
        x_list.append(x)
        y_list.append(y)
        
    return x_list, y_list

def plot_trajectory(x_list, y_list):
    plt.plot(x_list,y_list)
    plt.xlabel("X - horizontal distance (m)")
    plt.ylabel("Y -vertical distance (m)")
    plt.axis('equal')
    plt.grid(True)
    plt.title("Trajectory of a ball under the effect of Earth's gravity only")
    plt.show()

# Taking imput for velocity, launch angle.
v0 = float(input("Enter the initial launch velocity in m/s²: "))
launch_angle = float(input("Enter the launch angle in degrees : "))
x_list,y_list = projectile_motion(v0,launch_angle)
plot_trajectory(x_list,y_list)
```
